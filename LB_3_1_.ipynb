```python
import matplotlib.pyplot as plt
import numpy as np

# Задаємо рівняння параболи у вигляді y = ax^2 + bx + c
a = 4
b = -3
c = 2

# Генеруємо значення x в діапазоні від -10 до 10
x = np.linspace(-10, 10, 100)

# Обчислюємо значення y за рівнянням параболи
y = a * x**2 + b * x + c

# Побудова графіку
plt.plot(x, y, label='Парабола: $y = ax^2 + bx + c$')

# Додаємо підписи до графіку
plt.title('Графік параболи')
plt.xlabel('x')
plt.ylabel('y')

# Додаємо легенду
plt.legend()

# Показуємо графік
plt.grid(True)
plt.show()
```
